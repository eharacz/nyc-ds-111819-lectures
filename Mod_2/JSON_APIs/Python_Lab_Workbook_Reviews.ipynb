```python
def rev_api_call(url, headers):
    response = requests.get(url, headers=headers)
    data = response.json()
    review = data['reviews']
    return review
```

```python
def rev_parse_data(reviews, business_id):
    parsed_revs = []
    for review in reviews:
        rev_tuple = (review['id'], review['text'], review['rating'], review['time_created'], business_id)
        parsed_revs.append(rev_tuple)
    return parsed_revs
```

```python
def rev_insert_func(cnx, cursor, parsed_data):
    stmt = "INSERT INTO reviews (review_id, review_text, rating, time_created, business_id) VALUES (%s, %s, %s, %s, %s)"
    cursor.executemany(stmt, parsed_data)
    cnx.commit()
    print('Check the workbench.')
```

```python
def rev_paginate(biz_id_list):
    for b in biz_id_list:
        url= f'https://api.yelp.com/v3/businesses/{b}/reviews'
        reviews = rev_api_call(url, headers)
        parsed_reviews = rev_parse_data(reviews, b)
        rev_insert_func(cnx, cursor, parsed_reviews)
```

```python
rev_paginate(biz_id_list)
```
